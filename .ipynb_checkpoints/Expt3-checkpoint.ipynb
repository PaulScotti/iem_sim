## Initializing the model

```python
with pm.Model() as hierarchical_model_within:
    # Priors (p stands for prior, 
    # I leave out the letter for the group-level parameters we can later make inferences on)
    target_mu = pm.TruncatedNormal('target_mu', mu=0, sd=.5, lower=np.deg2rad(-15), upper=np.deg2rad(15))
    target_mu_sd = pm.Gamma('target_mu_sd', alpha=2, beta=10)
    target_mu_sd_sd = pm.Gamma('target_mu_sd_sd', alpha=2, beta=10)
    
    p_memA = pm.Exponential('p_memA', .1)
    p_memB = pm.Exponential('p_memB', .1)
    
    p_tarA = pm.Exponential('p_tarA', .1)
    p_tarB = pm.Exponential('p_tarB', .1)
    
    p_target_shape = pm.Gamma('p_target_shape', 2, .1)
    p_target_rate = pm.Gamma('p_target_rate', 2, .1)
    p_swap_shape = pm.Gamma('p_swap_shape', 2, .1)
    p_swap_rate = pm.Gamma('p_swap_rate', 2, .1)
    
    # Deterministic transformations   
    mem = pm.Deterministic('mem',p_memA/(p_memA + p_memB)) # 1-mem = proportion of random guessing
    target = pm.Deterministic('target',mem * (p_tarA/(p_tarA + p_tarB)))
    swap = pm.Deterministic('swap',mem * (1-(p_tarA/(p_tarA + p_tarB)))) 
    target_k = pm.Bound(pm.Gamma, lower=.609, upper=33.33)('target_k', alpha=p_target_shape, beta=p_target_rate)
    swap_k = pm.Bound(pm.Gamma, lower=.609, upper=33.33)('swap_k', alpha=p_swap_shape, beta=p_swap_rate)

    # Individual subject level (i stands for individual-level parameter)
    i_target_mu_tilde = pm.Normal('i_target_mu_tilde', mu=0, sd=1, shape=nsubj) #non-centered reparameterization
    i_target_mu = pm.Deterministic('i_target_mu', target_mu + i_target_mu_tilde * target_mu_sd)
    i_mem = pm.Beta('i_mem', alpha=p_memA, beta=p_memB, shape=nsubj)
    i_tar = pm.Beta('i_tar', alpha=p_tarA, beta=p_tarB, shape=nsubj) # tar refers to conditional probability of target
    i_target_k = pm.Bound(pm.Gamma, lower=.609, upper=33.33)('i_target_k', alpha=p_target_shape, beta=p_target_rate, shape=nsubj)
    i_swap_k = pm.Bound(pm.Gamma, lower=.609, upper=33.33)('i_swap_k', alpha=p_swap_shape, beta=p_swap_rate, shape=nsubj)
    
    # Trial-level
    dists = [
        pm.Uniform.dist(lower=np.full(ntrials,-np.pi), upper=np.full(ntrials,np.pi)), #random guessing
        pm.VonMises.dist(mu=i_target_mu[id], kappa=i_target_k[id]), #target with bias
        pm.VonMises.dist(mu=swap_loc, kappa=i_swap_k[id]), #swap
    ]
    
    w = tt.stack([1 - i_mem[id], i_mem[id] * i_tar[id], i_mem[id] * (1-i_tar[id])])
    
    y = pm.Mixture('y', w.T, dists, observed=color_err)
    
```

**I see warnings, does that mean my model is wrong?**

Above we see two warnings: divergences and effective sample size:

*Divergences*: 

Divergences in HMC occur when the sampler finds itself in regions of extremely high curvature (such as the opening of the a hierarchical funnel). Broadly speaking, the sampler is prone to malfunction in such regions, causing the sampler to fly off towards to infinity. The ruins the chains by heavily biasing the samples. (from [George Ho, PyMC core developer]) 

The PyMC3 ["Diagnosing biased inference with divergences"] page states that if your divergences are false positives, then your trace plots shouldn't show any stickyness. If your infrequent divergences do not seem concentrate anywhere in parameter space, this is indicative of the divergences being false positives. 

They recommend that if you have divergences then you should try to use [non-centered parameterization], which you see is implemented above for the mushift parameter. A non-centered implementation allows the Markov chains to explore deep into the neck of the hierarchical funnel.

*Effective sample size*: 

A low number of effective samples is usually an indication of strong autocorrelation in the chain. You can plot the trace autocorrelation or look at the trace plot directly to check if the sampler is either getting stuck or not exploring enough. Usually, this could be improved by using a more efficient sampler like NUTS, but here we already are using NUTS. You could try to alter the step size to improve effective sample size (by altering target_accept), but note that too small a step size will accept all proposals (i,e., random walk behaviour) and too large a step size will get your chains stuck. PyMC3's default target_accept is .80 but [George Ho] recommends a default target_accept of .90.

Also note that PyMC3 might be too sensitive with their effective sample size warnings. STAN's warning for effective sample size only shows when the n_eff for a parameter is less than 1% of the sample size! (paraphrased from [discourse.pymc.io])


[George Ho, PyMC core developer]: https://eigenfoo.xyz/bayesian-modelling-cookbook/#fixing-divergences
["Diagnosing biased inference with divergences"]: https://docs.pymc.io/notebooks/Diagnosing_biased_Inference_with_Divergences.html
[non-centered parameterization]: https://twiecki.github.io/blog/2017/02/08/bayesian-hierchical-non-centered/

## Checking/validating the model

So first off we can summarize all the results using the pm.summary() command. We have too many parameters though so we need to specify which ones we want to examine by using the "var_names" argument. Feel free to examine the other parameters and see what the 95% HDI (highest density interval aka HPD or highest posterior density) is. 95% HDIs indicate that the true parameter value has a 95% probability of lying within this interval. Values outside the intervals may be considered sufficiently implausible to be untenable (Lindley, 1965). 

We could have used a ROPE (Region of Practical Equivalence; Kruschke, 2014), where if the HDI is outside the ROPE then it's a credible difference but if the HDI is inside the ROPE then this supports the null. We didn't use this approach because we did not preregister a ROPE (nor did we come up with a ROPE before analyzing the data) so I felt like adopting the ROPE approach would be too susceptible to improper research practices.

### Gelman-Rubin diagnostic

Tests for lack of convergence by comparing the variance between multiple chains to the variance within each chain. If convergence has been achieved, the between-chain and within-chain variances should be identical. Gelman et al. (2004) recommend R to be less than 1.1. You also get this output using pm.summary().

Gelman, A., Carlin, J. B., Stern, H. S., and Rubin, D. B. (2004). Bayesian Data
Analysis. Chapman & Hall/CRC, Boca Raton, FL,.

```python
pm.diagnostics.gelman_rubin(posterior,var_names=["target_mu","target","swap"])
```

### Geweke diagnostic

Return z-scores for convergence diagnostics. Compare the mean of the first % of series with the mean of the last % of series. x is divided into a number of segments for which this difference is computed. If the series is converged, this score should oscillate between -1 and 1. If say, the first 10% of the plot is significantly different from the rest of the plot, this might indicate a longer burn-in is necessary. Note that we could do burn-in by doing something like posterior = posterior[500:] to discard the first half of all samples across chains, but it looks like we don't need to do this and that the tuning phase of the sampler already allowed for adequate mixing and initialization of the sampler. 

```python
plt.plot(pm.geweke(posterior['target_mu'])[:,1], 'o')
plt.axhline(1, c='red'); _ = plt.axhline(-1, c='red')
plt.gca().margins(0.05)
```

```python
pm.summary(posterior,var_names=["target","swap","target_mu","target_k","swap_k"])
# target: proportion of responses attributed to target distribution
# swap: proportion of responses attributed to swap distribution
# random_guess: proportion of responses attributed to random guessing distribution (equal to 1-target-swap)
# target_mu: center of target distribution (negative=repulsion, positive=attraction)
# target_k and swap_k: width of the target and swap distributions, in terms of von Mises concentration (kappa) 
```

```python
# functions to convert between the von Mises concentration (kappa) and standard deviation
# this way we can always use standard deviation when dealiing with neural or voxel bandwidths

k2sd = lambda k : np.sqrt( -2 * np.log( sp.special.iv(1,k) / sp.special.iv(0,k) ) )

def sd2k(S):
    try:
        len(S)
    except:
        S = np.array([S])
    R = np.exp(-S**2/2)
    K = 1/(R**3 - 4 * R**2 + 3 * R)
    if np.any(R < 0.85):
        K[R < 0.85] = -0.4 + 1.39 * R[R < 0.85] + 0.43/(1 - R[R < 0.85])
    if np.any(R < 0.52):
        K[R < 0.52] = 2 * R[R < 0.53] + R[R < 0.53]**3 + (5 * R[R < 0.53]**5)/6
    if len(S) == 1:
        K = float(K)
    return K
```

```python
# prep publication-quality table:
# convert kappa to sd and radian to degrees, 
# calculate random guessing proportion, and output pandas dataframe
var_names = ["target","swap","random_guess","target_mu","target_k","swap_k"]
table = np.full( (len(var_names), 3), np.nan)
for var in np.arange(len(var_names)):
    if var == 2:
        table[var,0] = 1 - table[0,0] - table[1,0];
        table[var,1] = pm.hpd(1-posterior[var_names[var-2]]-posterior[var_names[var-1]])[0]
        table[var,2] = pm.hpd(1-posterior[var_names[var-2]]-posterior[var_names[var-1]])[1]
    else:
        table[var,0] = np.mean(posterior[var_names[var]])
        table[var,1] = pm.hpd(posterior[var_names[var]])[0]
        table[var,2] = pm.hpd(posterior[var_names[var]])[1]
        if var_names[var] == "target_mu":
            table[var,:] = np.rad2deg(table[var,:])
        elif var_names[var] == "target_k" or var_names[var] == "swap_k":
            table[var,0] = np.rad2deg(k2sd(np.mean(posterior[var_names[var]])))
            table[var,1] = np.rad2deg(k2sd(pm.hpd(posterior[var_names[var]])[1]))
            table[var,2] = np.rad2deg(k2sd(pm.hpd(posterior[var_names[var]])[0]))
```

```python
# show table:
expt2_table = pd.DataFrame(table,
             index=["Target proportion","Swap proportion","Random guess proportion","Target shift (deg.)","Target SD (deg.)","Swap SD (deg.)"],
             columns=['Mean','HDI 2.5','HDI 97.5']) 
expt2_table

## convert table to LaTeX for use in a paper:
# print(model_table.to_latex(float_format="%.3f"))
```

### Model comparison

Below we compare our three models: full model, model without target shift, and model without swaps. We estimate how well a model fit the data while penalizing complexity using the measures of WAIC and PSIS-LOO.

WAIC (Watanabe 2010) is a fully Bayesian criterion for estimating out-of-sample expectation, using the computed log pointwise posterior predictive density (LPPD) and correcting for the effective number of parameters to adjust for overfitting.

LOO cross-validation is an estimate of the out-of-sample predictive fit. In cross-validation, the data are repeatedly partitioned into training and holdout sets, iteratively fitting the model with the former and evaluating the fit with the holdout data. Vehtari et al. (2016) introduced an efficient computation of LOO from MCMC samples, which are corrected using Pareto-smoothed importance sampling (PSIS) to provide an estimate of point-wise out-of-sample prediction accuracy.

```python
# load posteriors without running models
posterior = pm.load_trace("Expt2/long/posterior",hierarchical_model)
posterior_inflexible_mu = pm.load_trace("Expt2/long/posterior_inflexible_mu",hierarchical_model_inflexible_mu)
# posterior_no_swap = pm.load_trace("Expt2/long/posterior_no_swap",hierarchical_model_no_swap)
```

```python
hierarchical_model.name = 'Full Model'
hierarchical_model_inflexible_mu.name = "No Target Shift Model"
hierarchical_model_no_swap.name = "No Swap Model"

waic_results = pm.compare({hierarchical_model: posterior, 
            hierarchical_model_inflexible_mu: posterior_inflexible_mu,
            hierarchical_model_no_swap: posterior_no_swap})

loo_results = pm.compare({hierarchical_model: posterior, 
            hierarchical_model_inflexible_mu: posterior_inflexible_mu,
            hierarchical_model_no_swap: posterior_no_swap}, ic="LOO")
```

```python
display(waic_results)
display(loo_results)
pm.compareplot(waic_results)
pm.compareplot(loo_results)
```

Above is a summary plot of WAIC and LOO results. Empty circle: WAIC/LOO value, where lower value indicates better fit. Black error bars are standard deviation. Vertical dashed line indicates lowest WAIC/LOO value of the possible models. Filled black dots are the in-sample deviance of each model. Gray triangle indicates the WAIC/LOO difference between the model and the best-fitting model, with gray error bar denoting standard error of the difference in WAIC/LOO values.

In other words, the gray error bar approximates the uncertainty of WAIC/LOO estimates. If the gray error bar does not overlap with the vertical dotted line (the best-fit model estimate), then this demonstrates that the model with the lowest WAIC/LOO estimate is the best-fitting model. In both cases, we observe that the Full Model best fit the data.

Note that in the WAIC table, var_warn = 1 for the Full Model and No Target Shift Model. This occurs if posterior variance of the log predictive densities exceeds 0.4 and suggests unreliable WAIC estimates (Vehtari, Gelman, & Gabry, 2016). Similarly, for the LOO table, shape_warn will equal 1 if the pareto distribution is greater than 0.7 for one or more samples, and this would also suggest unreliable estimates. The LOO table showed no such warnings so we will interpret model comparison using PSIS-LOO (although the interpretations are the same either way).

Overall, this model comparison tells us that the shift and swap parameters explained nontrivial variance in the memory response distribution.
